```python
## Using reset to initial position when empty
class Queue: ## state of queue: head->position to delete, tail->position to insert
         ## when head == tail there are 2 states empty as well as full
    def __init__(self,n):
        self.array = [ None for i in range(n)]
        self.head =-1 ## next position to remove
        self.tail=0 ## next position to insert
        self.size = n
    def is_full(self):
        return True if self.tail == self.head else False      
    def is_empty(self):
        return True if self.head == -1 else  False        
    def enqueue(self,data):
        if self.is_full():
            print ("Full")
            return
        else:
            self.array[self.tail] = data
            self.tail = (self.tail+1)%self.size
            if self.head == -1: #1st enqueue
                self.head=0
    def dequeue (self):
        if self.is_empty():
            print("Empty")
            return 
        else:
            ret=self.array[self.head]
            self.head = (self.head+1)%self.size
            if self.head == self.tail: # empty list
                self.head = -1 ## reset
                self.tail = 0
            return ret
    def print(self):
        return [ i for i in self.array]
```

```python
## using an empty slot
class Queue():
    def __init__(self,size):
        self.head=0
        self.tail=0
        self.size=size+1
        self.array = [None for i in range(size+1)] #using an empty slot to check for full
    def isEmpty(self):
        return self.head == self.tail
    def isFull(self):
        return self.head == (self.tail+1)%self.size
    def enqueue(self,data):
        if self.isFull():
            print("Full")
        else:
            self.array[self.tail]=data
            self.tail  = (self.tail+1)%self.size
    def dequeue(self):
        if self.isEmpty():
            print("Empty")
        else:
            res=self.array[self.head]
            self.head = (self.head+1)%self.size
            return res
    def print(self):
        for i in  range(self.head, self.tail):
            print (self.array[i], end=",")
        print()
```

```python
## Using count
class Queue():
    def __init__(self,size):
        self.head=0 ## both empty and full condition
        self.tail=0
        self.size=size
        self.array = [None for i in range(size)]
        self.count=0
    def isEmpty(self):
        return self.count == 0
    def isFull(self):
        return self.count == self.size
    def enqueue(self,data):
        if self.isFull():
            print("Full")
        else:
            self.array[self.tail]=data
            self.tail  = (self.tail+1)%self.size
            self.count+=1
    def dequeue(self):
        if self.isEmpty():
            print("Empty")
        else:
            res=self.array[self.head]
            self.head = (self.head+1)%self.size
            self.count-=1
            return res
    def print(self):
            cur = self.head
            i = 0
            while i<self.count:
                print(self.array[cur],end=";")
                i+=1
                cur = (cur+1)%self.size
            print()
```
